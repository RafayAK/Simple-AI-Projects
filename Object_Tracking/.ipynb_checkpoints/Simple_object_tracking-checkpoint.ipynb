# Simple object tracking with OpenCV using centroid tracking

This object tracking algorithm uses _centroid tracking_, it relies on Euclidean distance between:
1. Existing object centroids (i.e. objects the centroid tracker has already seen before)
2. New object centroid between subsequent frames in a video

## Centroid Tracking Algorithm

__Step #1: Accept bounding box coordinates and compute centroids__

<img src="https://www.pyimagesearch.com/wp-content/uploads/2018/07/simple_object_tracking_step1.png" alt="Fig.1">

The centroid tracking algorithm assumes that we are passing in a set of bounding box (x, y)-coordinates for each detected object in ***every single frame.***

These bounding boxes can be produced by any type of object detector you would like (color thresholding + contour extraction, Haar cascades, HOG + Linear SVM, SSDs, Faster R-CNNs, etc.), provided that they are computed for every frame in the video

Once we have the bounding box coordinates we must compute the “centroid”, or more simply, the center (x, y)-coordinates of the bounding box. Figure above demonstrates accepting a set of bounding box coordinates and computing the centroid.

Since these are the first initial set of bounding boxes presented to our algorithm we will assign them unique IDs.

__Step #2: Compute Euclidean distance between new bounding boxes and existing objects__

<img src="https://www.pyimagesearch.com/wp-content/uploads/2018/07/simple_object_tracking_step2.png">

For every subsequent frame in our video stream we apply __Step #1__ of computing object centroids; however, instead of assigning a new unique ID to each detected object (which would defeat the purpose of object tracking), we first need to determine if we can associate the new object centroids (yellow) with the old object centroids (purple). To accomplish this process, we compute the Euclidean distance (highlighted with green arrows) between each pair of existing object centroids and input object centroids.

From Figure above you can see that we have this time detected three objects in our image. The two pairs that are close together are two existing objects.

We then compute the Euclidean distances between each pair of original centroids (yellow) and new centroids (purple). But how do we use the Euclidean distances between these points to actually match them and associate them?

__Step #3: Update (x, y)-coordinates of existing objects__

<img src="https://www.pyimagesearch.com/wp-content/uploads/2018/07/simple_object_tracking_step3.png">

The primary assumption of the centroid tracking algorithm is that a given object will potentially move in between subsequent frames, but the distance between the centroids for frames $F_t$ and $F_{t + 1}$ will be smaller than all other distances between objects.

Therefore, if we choose to associate centroids with minimum distances between subsequent frames we can build our object tracker.

In Figure above you can see how our centroid tracker algorithm chooses to associate centroids that minimize their respective Euclidean distances.

But what about the lonely point in the bottom-left?

It didn’t get associated with anything — what do we do with it?


__Step #4: Register new objects__

<img src="https://www.pyimagesearch.com/wp-content/uploads/2018/07/simple_object_tracking_step4.png">

In the event that there are more input detections than existing objects being tracked, we need to register the new object. “Registering” simply means that we are adding the new object to our list of tracked objects by:

1. Assigning it a new object ID
2. Storing the centroid of the bounding box coordinates for that object We can then go back to __Step #2__ and repeat the pipeline of steps for every frame in our video stream.

Figure above demonstrates the process of using the minimum Euclidean distances to associate existing object IDs and then registering a new object.

__Step #5: Deregister old objects__
Any reasonable object tracking algorithm needs to be able to handle when an object has been lost, disappeared, or left the field of view.

Exactly how you handle these situations is really dependent on where your object tracker is meant to be deployed, but for this implementation, we will deregister old objects when they cannot be matched to any existing objects for a total of ___N subsequent frames___.


## Lets code up the above steps

```python
from scipy.spatial import distance as dist
from collections import OrderedDict
import numpy as np

class CentroidTracker:
    def __init__(self, maxDisappeared=50):
        # initalze the next unique object ID along with two ordered
        # dictionaries used to keep track of the mapping a given
        # object ID to its centroid and number of consecutive frames it
        # has been marked as "disappeared", respectively
        
        self.nextObjectID = 0 # counter variable to assign unique IDs
        self.objects = OrderedDict() # this dict stores ID->centroid(x,y)
        self.disappeared = OrderedDict() # maintains num of consecutive frames a particual object ID has been missing
        
        # store the nunmber of maximum consecutive frames a given 
        # object is allowed to be marked as "disappeared" untill we
        # need to deregister the object from tracking
        self.maxDisappeared = maxDisappeared 
        
    
    # register() method is responsible for adding new objects to our tracker
    def register(self, centroid):
        # when registering an object we use the next available ID
        # to store centroid
        self.objects[self.nextObjectID] = centroid
        self.disappeared[self.nextObjectID] = 0
        self.nextObjectID += 1 # increment ID to a new value
        
    # deregister() method is responsible for removing objects that 
    # have disapperaed
    def deregister(self, objectID):
        # to deregister an object ID we delete the object ID from
        # both the dictionaries 
        del self.objects[objectID]
        del self.disappeared[objectID]
        
    
    def update(self, rects):
        # a 'rects' contains a list of tuples representing rectangles, 
        # rectangle->(startX, startY, endX, endY)
        
        
        # check to see if the list of input bounding box rectangles 
        # is empty:
        if len(rects) == 0:
            # loop over any existing tracked objects and mark them as
            # disappeared
            for objID in self.disappeared.keys():
                self.disappeared[objID] +=1
                
                # if we have reached a max number of consecutive
                # frames where a given object has been marked as 
                # missing, deregister it
                if self.disappeared[objID] == self.maxDisappeared:
                    self.deregister(objID)
            
            # return early as there are no centroid or tracking info
            # to update
            return self.objects
        
        # initaialze an array of input centroids for the current frame
        inputCentroids = np.zeros((len(rects), 2), dtype="int")
        
        # loop over the bounding box rectangles to calculate each
        # rectangle's center
        for (i, (startX, startY, endX, endY)) in enumerate(rects):
            cX = int((startX + endX)/2.0)
            cY = int((startY + endY)/2.0)
            inputCentroids[i] = (cX, cY)
        
        # If we are currently not tracking any objects take input
        # centroids and register each of them
        if len(self.objects) == 0:
            if i in range(len(inputCentroids)):
                self.register(inputCentroids[i])
        
        # otherwise, we need to try and match the input centroids to
        # existing object centroids
        else:
            # grab the set of all object IDs and corresponding 
            # centroids we are currently tracking
            objectIDs = list(self.objects.keys())
            objectCentroids = list(self.objects.values())
            
            # compute the distance between each pair of object
            # centroid and input centroids, respectively -- goal
            # is to match an input centroid to an existing object
            # centroid
            D = dist.cdist(np.array(objectCentroids), inputCentroids) # calculates euclidean
            
            # inorder to perform this matching we must
            # 1- find the smallest value in each row and then
            # 2- sort the row indexs based on their minimum values
            #    so that the row with the smallest value is at the
            #    front of the index list
            rows = D.min(axis=1).argsort()
            
            # next, we perform a similar process on the columns by
            # finding the smallest value in each column and then
            # sorting using the previously computed row index list
            cols = D.argmin(axis=1)[rows]
            
            # in order to determine if we need to update, register,
            # or deregister an object we need to keep track of which
            # of the rows and column indexes we have already examined
            usedRows = set() # initialize set
            usedCols = set() # initialize set
            
            # loop over combo of (row, column) index tuples
            for (row, col) in zip(rows, cols):
                # if we already examined either the row or 
                # column value before ignore it
                if row in usedRows or col in usedCols:
                    continue
                
                # otherwise, grab the object ID for the current row,
                # set its new centroid, reset the dissappeared
                # counter
                objectID = objectIDs[row]
                self.objects[objectID] = inputCentroids[col]
                self.disappeared[objectID] = 0
                
                # indicate that we have examined each of the row and 
                # column index, respectively
                usedRows.add(row)
                usedCols.add(col)
            
            # compute both the row and column index we have NOT yet
            # examined
            unusedRows = set(range(0, D.shape[0])).difference(usedRows)
            unusedCols = set(range(0, D.shape[1])).difference(usedCols)
            
            
            # in the event that the number of object centroids is 
            # equal or greater than the number of input centroids
            # we need to check and see if some of these objects have
            # potentially disappeared
            if D.shape[0] >= D.shape[1]:
                # loop over the unused row indexes
                for row in unusedRows:
                    # grab the object ID for the corresponding row
                    # index and increment the disappeared counter
                    objectID = objectIDs[row]
                    self.disappeared[objectID] += 1
                    
                    # check to see if the number of consecutive 
                    # frames the object has been marked "disappeared"
                    # for warrants deregistering the object
                    if self.disappeared[objectID] > self.maxDisappeared:
                        self.deregister(objectID)
            else:
                # otherwise if the number of input centroids is greater 
                # than the number of existing object centroids we need
                # to register each new input centroid as a trackable
                # object
                for col in unusedCols:
                    self.register(inputCentroids[col])
            
        # return set of trackable objects    
        return self.objects

            
```

## Understanding the centroid tracking distance relationship 

```python
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(42)

print("Object centroids: ")
objectCentroids = np.random.uniform(size=(2,2))
print(objectCentroids)
print("\n\n")

print("New Centroids:")
centroids = np.random.uniform(size=(3,2)) 
print(centroids)
```

Let's plot these.


_The red 'x' marks represent the objects **tracked from the previous frame**_.

_The yellow 'o' marks represent the **new objects** found_

```python
plt.figure(1)
plt.style.use('classic')
plt.grid()
# plt.xlim([-2.0, 2.0])
# plt.ylim([-2.0, 2.0])
plt.xlabel('$x$', size=20)
plt.ylabel('$y$', size=20)
plt.title('Input 2D points', size=18)
plt.scatter(x=objectCentroids[:, 0], y=objectCentroids[:, 1], s=50, marker="x" ,c='r')
plt.scatter(x=centroids[:, 0], y=centroids[:, 1], s=50, c='yellow')
plt.show()
```

__Form here we can compute the Euclidean distance of all the 'x' points to 
all the yellow points__

```python
D = dist.cdist(objectCentroids, centroids) # calculate eclidean

print("Eucludiean distances:")
print(D)
```

```python
D.shape
```

The result is a matrix **D**  of ___distances with two rows (# of existing object centroids) and three columns (# of new input centroids).___

Lets' plot and visualize these:

```python
plt.figure(2)
plt.style.use('classic')
plt.grid()
# plt.xlim([-2.0, 2.0])
# plt.ylim([-2.0, 2.0])
plt.xlabel('$x$', size=20)
plt.ylabel('$y$', size=20)
plt.title('Input 2D points', size=18)
plt.scatter(x=objectCentroids[:, 0], y=objectCentroids[:, 1], s=200, marker="x" ,c='r')
plt.scatter(x=centroids[:, 0], y=centroids[:, 1], s=50, c='yellow')

def connectpoints(x,y,p1,p2, euclid_dist, color):
    plt.plot([x,p1], [y,p2], color=color)
    # calculate where to dsipaly distance
    half_x_dist = (x+p1)/2
    half_y_dist = (y+p2)/2
    plt.text(half_x_dist,half_y_dist,"{:.3f}".format(euclid_dist))


color = ['c', 'm'] # cyan and megnta   
for i in range(len(objectCentroids)):
    for j in range(D.shape[1]): #Go over all eclideans
        connectpoints(objectCentroids[i, 0], objectCentroids[i, 1], 
                      centroids[j, 0], centroids[j, 1], 
                      euclid_dist=D[i,j], color= color[i])
    
```

We can now see which of the _yellow 'o'_ marks is closest to the _red 'x'_.

Time extract minimum distances

First, we find the minimum value for each row, allowing us to figure out which existing object is closest to the new input centroid .

```python
D.min(axis=1) # returns min 'value' along axis
```

Then sorting on these values we can obtain the indexes of these rows

```python
rows = D.min(axis=1).argsort() # argsort Returns the indices that would sort an array.
```

```python
rows
```

In this case, the second row (index 1 ) has the smallest value and then the first row (index 0 ) has the next smallest value.

Using a similar process for the columns:

```python
D.argmin(axis=1) # Returns the 'indices' of the minimum values along an axis.
```

we first examine the values in the columns and find the index of the value with the smallest column, as above.

We then sort these values using our existing rows

```python
cols = D.argmin(axis=1)[rows]
cols
```

Let’s print the results and analyze them:

```python
 print(list(zip(rows, cols)))
```

Analyzing the results, we find that:

- ***D[1, 2]  has the smallest Euclidean distance implying that the second existing object(1) will be matched against the third input(2) centroid.***
- ***And D[0, 1]  has the next smallest Euclidean distance which implies that the first(0) existing object will be matched against the second input(1) centroid.***
- ***The unmatched centroid is a new object***

Let's visualize:

```python
plt.figure(3)
plt.style.use('classic')
plt.grid()
# plt.xlim([-2.0, 2.0])
# plt.ylim([-2.0, 2.0])
plt.xlabel('$x$', size=20)
plt.ylabel('$y$', size=20)
plt.title('Input 2D points', size=18)
plt.scatter(x=objectCentroids[:, 0], y=objectCentroids[:, 1], s=200, marker="x" ,c='r')
plt.scatter(x=centroids[:, 0], y=centroids[:, 1], s=50, c='yellow')

def connectpoints(x,y,p1,p2, euclid_dist, color):
    plt.plot([x,p1], [y,p2], color=color)
    # calculate where to dsipaly distance
    half_x_dist = (x+p1)/2
    half_y_dist = (y+p2)/2
    plt.text(half_x_dist,half_y_dist,"{:.3f}".format(euclid_dist))


color = ['c', 'm'] # cyan and megnta   
for (row, col) in zip(rows, cols):
    connectpoints(objectCentroids[row, 0], objectCentroids[row, 1], 
                  centroids[col, 0], centroids[col, 1], 
                  euclid_dist=D[row,col], color= color[row])
    
```

## Now we can implement the object tracking script

We will use the Resnet based Caffe model from the DNN_Models directory for face detection

```python
#import necessary packages
#from pyimagesearch.centroidtracker import CentroidTracker
from imutils.video import VideoStream
import numpy as np
import imutils
import time
import cv2

# define paths to model and its prototxt
model_architecture_path = "..\\DNN_MODELS\\opencv_face_detector.prototxt"
caffe_model_path = '..\\DNN_MODELS\\opencv_face_detector.caffemodel'

# define defalut confidence i.e. probabliblity to filter weak detections
confidence = 0.5
```

```python
# initialize our centroid tracker and frame dimensions 
ct = CentroidTracker()
(H, W) = (None, None)

# load model
print("[INFO] loading model...")
net = cv2.dnn.readNetFromCaffe(model_architecture_path, caffe_model_path)

# initialize video stream
print("[INFO] starting video stream from webcam...")
vs = VideoStream(src=0).start()
time.sleep(2.0) # sleep to let frames buffer
```

```python

# only for running in jupyter
from IPython.display import display , Image, clear_output
```

```python
# loop over the frames from the video stream
while True:
    # read the next frame from the webcam and resize it
    frame = vs.read()
    frame = imutils.resize(frame, width=400) # resize to w=400, maintain asptect ratio
    
    # if the frame dimensions are None, grab them 
    if W is None or H is None:
        (H, W) = frame.shape[:2] # (H, W, depth)
        
    # construct blog from the frame, pass it through the CNN
    # obtain our output predictions, and initialize the list of
    # bounding boxes
    blob = cv2.dnn.blobFromImage(image=frame, scalefactor=1.0, 
                             size=(H,W), mean=(104.0, 177.0, 123.0))
    net.setInput(blob)
    detections = net.forward()
    rects = []
    
    # loop over the detections
    for i in range(0, detections.shape[2]):
        # filter weak detections
        if detections[0,0,i,2] > confidence:
            box = detections[0,0,i, 3:7] * np.array([W, H, W, H])
            rects.append(box.astype("int"))
            
            # draw a bounding box surrounding the object so we can 
            # visualize it
            (startX, startY, endX, endY) = box.astype("int")
            cv2.rectangle(frame, (startX,startY), (endX,endY),
                          (0,255, 0), 2)
    
    # update the centroid tracker using computed set of bounding
    # boxes
    objects = ct.update(rects)
    
    # loop over the tracked objects
    for (objectID, centroid) in objects.items():
        # draw the ID of the object and the centroid of the
        # object on the frame
        text = "ID {}".format(objectID)
        # draw text just over the centroid
        cv2.putText(frame, text, org=(centroid[0] - 10, centroid[1] - 10),
                    fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=0.5, 
                    color=(0, 255,0), thickness=2)
        cv2.circle(frame, (centroid[0], centroid[1]), 4, (0, 255, 0),
                   -1) # thickness=-1 -> filled circle
        
    # show output
    # for jupyter first we need to save the image then dsplay it 
    # for a video like effect
    cv2.imwrite('pic.jpg', frame) # first we save it
    display(Image("pic.jpg"))
    
    clear_output(wait=True)
    
    key = cv2.waitKey(1) & 0xFF
    # if the `q` key was pressed, break from the loop
    if key == ord("q"):
        break
        
```

```python
vs.stop()
```
